```python
from scipy import misc, ndimage
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import skimage.transform  as ski
import skimage.transform as sk

from skimage import data
from skimage import transform as tf
import math

from __future__ import print_function

import skimage
from skimage import io, color
from skimage.morphology import disk
from skimage.morphology import dilation
from skimage.morphology import erosion
from skimage.morphology import diamond
from skimage.morphology import opening
from skimage.morphology import closing
from skimage.transform import rescale
```

```python
# Funcao para plotar uma imagem qualquer
def show(img):
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    ax.imshow(img, cmap=plt.cm.gray)
    ax.set_axis_off()
    plt.show()
```

Abrindo Imagem Preto e Branco escolhido

```python
#imagem preto e branca escolhida
img = io.imread('Hobbit.jpeg', as_grey=True)  
show(img)
```

### Erosão e Dilatação

```python
def dilatacao(N):
    out = dilation(img, skimage.morphology.square(N))
    show(out)
    
def erosao(N):
    out = erosion(img, skimage.morphology.square(N))
    show(out)
    
def dilatacao_erosao(N):
    out = erosion(img, skimage.morphology.square(N))
    out= dilation(out,skimage.morphology.square(N))
    show(out)

def erosao_erosao(N):
    out = erosion(img, skimage.morphology.square(N))
    out= erosion(out,skimage.morphology.square(N))
    show(out)
```

```python
print("Dilataçao:" )
dilatacao(5)
print("Erosao: ")
erosao(5)
print("Dilatacao + erosao: ")
dilatacao_erosao(5)
print("Erosao + erosao: ")
erosao_erosao(5)
```

### Fechamento e Abertura

```python
def abertura(N): 
    out = opening(img, skimage.morphology.square(7))
    show(out)
def fechamento(N):
    out = closing(img, skimage.morphology.square(7))
    show(out)
def abertura_fechamento(N):
    out = opening(img, skimage.morphology.square(N))
    out = closing(out, skimage.morphology.square(N))
    show(out)
    
```

```python
print("Abertura")
abertura(7)
print("Fechamento")
fechamento(7)
print("Abertura + Fechamento")
abertura_fechamento(7)
```

```python
print("Transladando a imagem")

shifted = ndimage.shift(img,[100, 50], cval=255)
plt.imshow(shifted, cmap='gray')
plt.show()

shifted = ndimage.shift(img,[100, 500], cval=0)
plt.imshow(shifted, cmap='gray')
plt.show()
```

```python

print("Escalonando a imagem")
scaled = misc.imresize(img,100)
plt.imshow(scaled, cmap='gray')
plt.show()

scaled = misc.imresize(img,500)
plt.imshow(scaled, cmap='gray')
plt.show()
```

```python
print("Cortando a imagem")
lx, ly = img.shape
crop_img = img[lx // 8: - lx // 8, ly // 8: - ly // 8]
plt.imshow(crop_img, cmap='gray')
plt.show()

lx, ly = img.shape
crop_img = img[lx // 5: - lx // 5, ly // 5: - ly // 5]
plt.imshow(crop_img, cmap='gray')
plt.show()

lx, ly = img.shape
crop_img = img[lx // 3: - lx // 3, ly // 3: - ly // 3]
plt.imshow(crop_img, cmap='gray')
plt.show()
```
