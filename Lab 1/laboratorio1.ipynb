```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import skimage.io as io
%matplotlib inline 
from matplotlib import pyplot as plt
import numpy as np
import skimage.exposure as skie
import seaborn as sns
```

```python
#carregando e mostrando a foto pretro e branco
img = mpimg.imread('yuukipb.jpg')      
plt.imshow(img)
plt.show()



```

```python
#carregando e mostrando a imagem colorida
img = mpimg.imread('toradoracolor.jpg')      
plt.imshow(img)
plt.show()
```

```python
#criando e mostrando o histograma da imagem preto e branco
img = io.imread('yuukipb.jpg')
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
#criando e mostrando o histograma da imagem colorida
img = io.imread('toradoracolor.jpg') 
color = [ 'r','g','b']
for i, c in enumerate(color) :
    plt.hist(img[:,:,i].flatten(),256, color=c)
plt.show()
```

```python
#funçao para mostar imagem e histograma 
def show(img):
    # Display the image.
    fig, (ax1, ax2) = plt.subplots(1, 2,
                                   figsize=(12, 3))

    ax1.imshow(img, cmap=plt.cm.gray)
    ax1.set_axis_off()

    # Display the histogram.
    ax2.hist(img.ravel(), lw=0, bins=256)
    ax2.set_xlim(0, img.max())
    ax2.set_yticks([])

    plt.show()
```

```python
#alterando a exposiçao do histograma

img = io.imread('yuukipb.jpg') 
#original
show(img)
#modificada
show(skie.rescale_intensity(img, in_range=(200,255), out_range=(0, 255)))
```

```python
#criando a imagem negativa invertendo o histograma
#imagem colorida
img = io.imread('toradoracolor.jpg') 
show(skie.rescale_intensity(img, in_range=(0,255), out_range=(255, 0)))
#imagem preto e branco
img = io.imread('yuukipb.jpg') 
show(skie.rescale_intensity(img, in_range=(0,255), out_range=(255,0)))
```
